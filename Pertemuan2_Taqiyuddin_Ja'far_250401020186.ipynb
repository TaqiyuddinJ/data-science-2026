```python
import pandas as pd
import numpy as np

#Muat dataset titanic
sumber = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(sumber)
df.head()
df.info()
```

```python
# Eksplorasi Awal
# Q1: Berapa jumlah baris dan kolom pada dataset Titanic? Kolom apa saja yang tersedia?
print(f"Jumlah baris dan kolom dalam format (baris,kolom): {df.shape}") # shape menjawab (baris,kolom)
print(f"Daftar kolom: {df.columns.tolist()}\n") #columns.tolist menampilkan daftar 12 kolom mulai dari 'PassengerId' sampai 'Embarked'
```

```python
# Q2: Kolom mana saja yang memiliki missing values? Berapa jumlah dan proporsinya (%) per kolom?
missing_data = df.isnull().sum()
print(missing_data) # pemeriksaan missing values. Jadi di kolom 'Age' ada 177, di 'Cabin' ada 687, di 'Embarked' ada 2
total_rows = len(df)
print(f"total penumpang = {total_rows}\n") # len(df) menghitung jumlah baris atau dalam data ini sebagai total populasi semua penumpang

missing_info = pd.DataFrame({
    'Jumlah Missing': missing_data,
    'Proporsi (%)': (missing_data / total_rows) * 100
})
print(missing_info[missing_info['Jumlah Missing'] > 0], "\n")
```

```python
# Q3: Berapa persen penumpang yang selamat (Survived=1)? Gunakan operasi NumPy atau Pandas.
percent_survived = df['Survived'].mean() * 100 # Menggunakan Pandas mean() dari seluruh kolom 'Survived',karena kolom 'Survived' berisi 0 dan 1
print(f"Persentase penumpang selamat: {percent_survived:.2f}%\n")
```

```python
# Filtering — Penumpang Wanita Kelas 1
# Q4: Filter penumpang wanita dari kelas 1 (Pclass=1). Berapa jumlahnya dan berapa rata-rata usianya?
female_class_1 = df[(df['Sex'] == 'female') & (df['Pclass'] == 1)] # Menggunakan Boolean Indexing dengan kondisi ganda
print(f"Jumlah penumpang wanita dari kelas 1: {len(female_class_1)}")
print(f"Rata-rata usia penumpang wanita dari kelas 1: {female_class_1['Age'].mean():.2f} tahun\n") # Menggunakan Pandas mean() dari seluruh kolom 'Age' dari female_class_1
```

```python
# Groupby & Agregasi
# Q5: Gunakan groupby untuk menghitung tingkat keselamatan (mean Survived) per kelas penumpang. Kelas mana yang paling tinggi?
survival_by_class = df.groupby('Pclass')['Survived'].mean()
print(survival_by_class)
print("\n")

hasil = df.groupby('Pclass').agg(
    total=('PassengerId','count'),
    selamat=('Survived','mean'),
).round(2)
print(hasil) # hasil aggregasi
print("\n")

print(f"Kelas dengan tingkat keselamatan tertinggi: {survival_by_class.idxmax()}\n")
```

```python
# Q6: Buat analisis tambahan berdasarkan variabel lain yang Anda anggap menarik. Sertakan interpretasi singkat.

# Analisis tingkat keselamatan berdasarkan pelabuhan keberangkatan (Embarked)
# C = Cherbourg, Q = Queenstown, S = Southampton

embarked_survival = df.groupby('Embarked')['Survived'].agg(['count', 'mean']).round(2).sort_values(by='mean', ascending=False)

print("Analisis Berdasarkan Pelabuhan Keberangkatan:")
print(embarked_survival)

# Interpretasi singkat:
# Penumpang yang naik dari Cherbourg (C) memiliki tingkat keselamatan tertinggi (sekitar 55%),
# sementara dari Southampton (S) memiliki tingkat keselamatan terendah meskipun jumlah penumpangnya paling banyak.
```
