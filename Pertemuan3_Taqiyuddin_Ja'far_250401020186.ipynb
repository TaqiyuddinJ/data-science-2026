```python
## Import semua library yang diperlukan: pandas, numpy, scipy.stats.mstats

import pandas as pd
import numpy as np
from scipy.stats import mstats
```

```python
## Muat dataset: df = pd.read_csv(url_atau_path_lokal)
df = pd.read_csv('housing_dirty.csv')

## Eksplorasi awal: df.info(), df.describe(), df.isnull().sum()
print(df.info())
print(df.describe())
print(df.shape)
print(df.isnull().sum())
```

```python
## Hapus baris duplikat: df.drop_duplicates(inplace=True)
# shape awal (130,7)
df.drop_duplicates(inplace=True)
print('Setelah hapus duplikat:', df.shape)
print((df.isnull().sum()/len(df)*100).round(2))
```

```python
## Normalisasi string: .str.strip().str.title() untuk kolom kota, .str.lower() untuk kondisi
df['kota'] = df['kota'].str.strip().str.title()
df['kondisi'] = df['kondisi'].str.lower()
```

```python
## Imputasi missing values: median untuk numerik, modus untuk kategorik
df['luas_m2'] = df['luas_m2'].fillna(df['luas_m2'].median())
df['harga_juta'] = df['harga_juta'].fillna(df['harga_juta'].median())
df['kamar'] = df['kamar'].fillna(df['kamar'].mode()[0])
```

```python
## Tangani outlier dengan IQR Fence pada kolom harga_juta dan luas_m2
for col in ['harga_juta', 'luas_m2', 'tahun_bangun']:
  Q1, Q3 = df[col].quantile([0.25, 0.75])
  IQR = Q3 - Q1
  df[col] = df[col].clip(Q1-1.5*IQR, Q3+1.5*IQR)
```

```python
## Validasi akhir: df.isnull().sum().sum() harus = 0, df.duplicated().sum() harus = 0
assert df.isnull().sum().sum() == 0, 'Masih ada missing!'
assert df.duplicated().sum() == 0, 'Masih ada duplikat!'
print('Shape akhir:', df.shape)
```

```python
## 10.Ekspor: df.to_csv('housing_clean.csv', index=False)
df.to_csv('housing_clean.csv', index=False)
print('Dataset bersih tersimpan!')
```
